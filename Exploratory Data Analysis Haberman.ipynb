```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
haberman = pd.read_csv("habermanE.csv")
```

```python
print(haberman.shape)
```

```python
print(haberman.columns)
```

```python
haberman["survStatus"].value_counts()
```

```python
haberman.plot(kind="scatter",x="age",y="opYear");
plt.show()
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(haberman,hue="survStatus",size=5).map(plt.scatter,"age","opYear").add_legend();
plt.show();
```

```python
plt.close();
sns.set_style("whitegrid");
sns.pairplot(haberman,hue="survStatus",size=3);
plt.show()
```

```python
h_surv=haberman.loc[haberman["survStatus"]==1];
h_dead=haberman.loc[haberman["survStatus"]==2];

plt.plot(h_surv["age"], np.zeros_like(h_surv['age']), 'o')
plt.plot(h_dead["age"], np.zeros_like(h_dead['age']), 'o')
plt.show();
```

```python
counts, bin_edges = np.histogram(h_surv['axilNodes'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)


counts, bin_edges = np.histogram(h_surv['axilNodes'], bins=20, 
                                 density = True)
pdf = counts/(sum(counts))
plt.plot(bin_edges[1:],pdf);

plt.show();
```
